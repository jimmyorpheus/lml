### 1. Import Data from a CSV File

* Using a local CSV File:
#### see: https://www-genesis.destatis.de/genesis/online/logon?language=de

```python
import pandas as pd

populationBundeslaenderPath = "./work/populationBundeslaenderUtf8.csv"

populationBundeslaender = pd.read_csv(populationBundeslaenderPath, sep=';')

fullTable = populationBundeslaender.iloc[4:21]

display(fullTable)


# Display Data from 2016 as Bar-Diagram

import matplotlib.pyplot as plt

popuBl2016 = populationBundeslaender.iloc[5:21,[0,10]]
popuBl2016.columns = ['Bundesländer', 'Bevölkerung']
popuBl2016['Bevölkerung'] = popuBl2016['Bevölkerung'].astype(int)
ax = plt.gca()
ax.ticklabel_format(style='plain')
display(popuBl2016)
display(popuBl2016.dtypes)
display(popuBl2016.columns)
popuBl2016.plot(x='Bundesländer', kind='bar', ax=ax)
```
